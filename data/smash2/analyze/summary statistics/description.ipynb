```python
import pandas as pd

# Liste des stocks
stocks = ['ASAI', 'CGAU', 'HL', 'RIOT']

# Dictionnaire pour stocker les DataFrames
dataframes = {}

# Charger les fichiers CSV et afficher les premières lignes et les statistiques résumées
for stock in stocks:
    file_path = f'/users/eleves-a/2022/janis.aiad/3A/EAP1/HFT_QR_RL/HFT_QR_RL/data/smash2/data/csv/{stock}/20240624.csv'
    df = pd.read_csv(file_path)
    dataframes[stock] = df
    
    # Afficher les premières lignes du DataFrame
    print(f"Premières lignes pour {stock}:")
    print(df.head())
    
    # Afficher les statistiques résumées du DataFrame
    print(f"Statistiques résumées pour {stock}:")
    print(df.describe())
    print("\n" + "="*80 + "\n")
```

```python
import matplotlib.pyplot as plt

# Liste des publisher_id
publisher_ids = [39, 40, 41]

# Tracer le mouvement des prix pour chaque stock et chaque publisher_id
for stock in stocks:
    df = dataframes[stock]
    
    for pub_id in publisher_ids:
        # Filtrer les données pour le publisher_id actuel
        df_filtered = df[df['publisher_id'] == pub_id]
        
        # Convertir ts_event en datetime
        df_filtered['ts_event'] = pd.to_datetime(df_filtered['ts_event'], unit='ns')
        
        # Créer une figure
        plt.figure(figsize=(10, 6))
        
        # Tracer le prix moyen
        plt.plot(df_filtered['ts_event'], df_filtered['price'], label='Mid Price', color='black')
        
        # Mettre à jour la mise en page
        plt.title(f"Mid Price for {stock} with publisher_id {pub_id}")
        plt.xlabel("Time")
        plt.ylabel("Price")
        plt.legend()
        
        # Afficher le graphique
        plt.show()
```
